```python
import pandas as pd


languages = ['en_US', 'de_DE', 'hi_IN', 'es_ES', 'fr_FR', 'it_IT', 'ar_SA', 'nl_NL', 'ja_JP', 'pt_PT']

def create_translation_pairs(df):
    pairs = []
    for _, row in df.iterrows():
        for src_lang in languages:
            for tgt_lang in languages:
                if src_lang != tgt_lang:

                    source_text = f"<{tgt_lang}> {row[src_lang]}"
                    target_text = row[tgt_lang]
                    pairs.append({"source": source_text, "target": target_text})
    return pd.DataFrame(pairs)
df_train = df_train.sample(frac=0.5, random_state=42)

train_pairs = create_translation_pairs(df_train)
test_pairs = create_translation_pairs(df_test)

print(f"Training pairs: {len(train_pairs)}, Testing pairs: {len(test_pairs)}")


train_pairs.to_csv("train_pairs.csv", index=False)
test_pairs.to_csv("test_pairs.csv", index=False)
```

```python
import sentencepiece as spm
from datasets import load_dataset

sp = spm.SentencePieceProcessor(model_file="mymodel.model")

def tokenize_function(example, max_length=100, lowercase=False):
    source_text = example["source"]
    target_text = example["target"]

    if lowercase:
        source_text = source_text.lower()
        target_text = target_text.lower()

    source_tokens = sp.encode(source_text, out_type=int)[:max_length]
    target_tokens = sp.encode(target_text, out_type=int)[:max_length]

    sos_token = sp.bos_id()
    eos_token = sp.eos_id()

    sos_token_str = sp.id_to_piece(sos_token)
    eos_token_str = sp.id_to_piece(eos_token)


    print(f"SOS token: {sos_token_str}, EOS token: {eos_token_str}")

    source_tokens = [sos_token] + source_tokens + [eos_token]
    target_tokens = [sos_token] + target_tokens + [eos_token]

    source_tokens_str = [sp.id_to_piece(token).replace("<s>", "<sos>").replace("</s>", "<eos>") for token in source_tokens]
    target_tokens_str = [sp.id_to_piece(token).replace("<s>", "<sos>").replace("</s>", "<eos>") for token in target_tokens]

    source_tokens_str = [sp.id_to_piece(token) for token in source_tokens]
    target_tokens_str = [sp.id_to_piece(token) for token in target_tokens]

    print(f"Source tokens: {source_tokens_str}")
    print(f"Target tokens: {target_tokens_str}")

    return {"source_tokens": source_tokens, "target_tokens": target_tokens}
```

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import sentencepiece as spm


class TranslationDataset(Dataset):
    def __init__(self, csv_file, sp_model):
        self.data = pd.read_csv(csv_file)
        self.sp_model = sp_model
        self.vocab_size = sp_model.vocab_size()

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        source_str = self.data.iloc[idx]["source_tokens"]
        target_str = self.data.iloc[idx]["target_tokens"]

        try:
            source_str = source_str.strip("[]")
            target_str = target_str.strip("[]")

            source = list(map(int, source_str.split()))
            target = list(map(int, target_str.split()))

            source = [min(token, self.vocab_size - 1) for token in source]
            target = [min(token, self.vocab_size - 1) for token in target]

            return torch.tensor(source, dtype=torch.long), torch.tensor(target, dtype=torch.long)

        except Exception as e:
            print(f"Error processing index {idx}: {e}")
            return None

def collate_fn(batch, pad_token=0):
    batch = [b for b in batch if b is not None]

    batch.sort(key=lambda x: len(x[0]), reverse=True)

    src_sequences, tgt_sequences = zip(*batch)

    src_padded = nn.utils.rnn.pad_sequence(src_sequences, batch_first=True, padding_value=pad_token)
    tgt_padded = nn.utils.rnn.pad_sequence(tgt_sequences, batch_first=True, padding_value=pad_token)

    return src_padded, tgt_padded

# Load your SentencePiece model
sp_model = spm.SentencePieceProcessor(model_file="mymodel.model")

print(f"Padding ID: {sp_model.pad_id()}, Vocab Size: {sp_model.vocab_size()}")

train_dataset = TranslationDataset("train_pairs_tokenized.csv", sp_model)
valid_dataset = TranslationDataset("test_pairs_tokenized.csv", sp_model)
test_dataset = TranslationDataset("test_pairs_tokenized.csv", sp_model)

# Define your data loaders here
train_iterator = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=lambda batch: collate_fn(batch, pad_token=sp_model.pad_id())
)

valid_iterator = DataLoader(
    valid_dataset,
    batch_size=32,
    shuffle=False,
    collate_fn=lambda batch: collate_fn(batch, pad_token=sp_model.pad_id())
)

test_iterator = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False,
    collate_fn=lambda batch: collate_fn(batch, pad_token=sp_model.pad_id())
)
```

```python
import torch.nn as nn
class LuongAttention(nn.Module):
    def __init__(self, hidden_dim):
        super(LuongAttention, self).__init__()
        self.attn = nn.Linear(hidden_dim, hidden_dim)

    def forward(self, decoder_hidden, encoder_outputs):
        energy = torch.bmm(encoder_outputs, decoder_hidden.unsqueeze(2)).squeeze(2)
        attn_weights = torch.softmax(energy, dim=1)
        context = torch.bmm(attn_weights.unsqueeze(1), encoder_outputs).squeeze(1)
        return context, attn_weights
```

```python
import torch.nn as nn
class Encoder(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_layers, dropout=0.5):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)

    def forward(self, src):
        embedded = self.embedding(src)
        outputs, (hidden, cell) = self.lstm(embedded)
        return outputs, hidden, cell

```

```python
class Decoder(nn.Module):
    def __init__(self, output_dim, embedding_dim, hidden_dim, num_layers, dropout=0.5):
        super(Decoder, self).__init__()
        self.embedding = nn.Embedding(output_dim, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        self.dropout = nn.Dropout(dropout)
        self.attention = LuongAttention(hidden_dim)

    def forward(self, tgt, hidden, cell, encoder_outputs):
        embedded = self.embedding(tgt).unsqueeze(1)
        output, (hidden, cell) = self.lstm(embedded, (hidden, cell))

        context, _ = self.attention(hidden[-1], encoder_outputs)

        output = torch.cat((output.squeeze(1), context), dim=1)
        prediction = self.fc(output)

        return prediction, hidden, cell
```

```python
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder, device):
        super(Seq2Seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device

    def forward(self, src, tgt, teacher_forcing_ratio=0.5):
        encoder_outputs, hidden, cell = self.encoder(src)

        batch_size = src.shape[0]
        tgt_len = tgt.shape[1]
        tgt_vocab_size = self.decoder.fc.out_features

        outputs = torch.zeros(batch_size, tgt_len, tgt_vocab_size).to(self.device)

        dec_input = tgt[:, 0]

        for t in range(1, tgt_len):
            output, hidden, cell = self.decoder(dec_input, hidden, cell, encoder_outputs)
            outputs[:, t, :] = output

            top1 = output.argmax(1)

            dec_input = tgt[:, t] if torch.rand(1).item() < teacher_forcing_ratio else top1

        return outputs
```

```python
import torch.optim as optim
import sentencepiece as spm
import torch

sp_model = spm.SentencePieceProcessor(model_file="mymodel.model")

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

INPUT_DIM = sp_model.vocab_size()
OUTPUT_DIM = sp_model.vocab_size()
EMBEDDING_DIM = 256
HIDDEN_DIM = 512
NUM_LAYERS = 2
DROPOUT = 0.5
LEARNING_RATE = 0.001

encoder = Encoder(INPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM, NUM_LAYERS, DROPOUT).to(device)
decoder = Decoder(OUTPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM, NUM_LAYERS, DROPOUT).to(device)
model = Seq2Seq(encoder, decoder, device).to(device)

criterion = nn.CrossEntropyLoss(ignore_index=sp_model.pad_id())
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
```

```python
import os
import torch

checkpoint_dir = "/content/drive/MyDrive/AI_Project/checkpoints"
resume_path = os.path.join(checkpoint_dir, "latest_checkpoint.pt")

# Check if the checkpoint exists
if os.path.exists(resume_path):
    print(f"Loading checkpoint from {resume_path}")
    checkpoint = torch.load(resume_path, map_location=torch.device("cuda" if torch.cuda.is_available() else "cpu"))
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    best_valid_loss = checkpoint['best_valid_loss']
    print(f"Resuming from epoch {start_epoch}")
else:
    print("No checkpoint found. Starting from scratch.")
```

```python
def train(model, iterator, optimizer, criterion, clip, teacher_forcing_ratio):
    model.train()
    epoch_loss = 0

    for i, (src, tgt) in enumerate(iterator):
        src = src.to(device)
        tgt = tgt.to(device)

        # Access vocab_size from sp_model instead of the model
        src = torch.clamp(src, max=sp_model.vocab_size()-1)
        tgt = torch.clamp(tgt, max=sp_model.vocab_size()-1)

        optimizer.zero_grad()

        output = model(src, tgt, teacher_forcing_ratio)

        output_dim = output.shape[-1]

        output = output[:, 1:].reshape(-1, output_dim)

        tgt = tgt[:, 1:].reshape(-1)

        loss = criterion(output, tgt)
        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), clip)

        optimizer.step()

        epoch_loss += loss.item()

        if i % 100 == 0:
            print(f"Batch {i}/{len(iterator)}, Loss: {loss.item():.4f}")

    return epoch_loss / len(iterator)
```

```python
def evaluate(model, iterator, criterion):
    model.eval()
    epoch_loss = 0

    with torch.no_grad():
        for i, (src, tgt) in enumerate(iterator):
            src = src.to(device)
            tgt = tgt.to(device)

            # Access vocab_size from sp_model instead of the model
            src = torch.clamp(src, max=sp_model.vocab_size()-1)
            tgt = torch.clamp(tgt, max=sp_model.vocab_size()-1)

            # teacher forcing of 0
            output = model(src, tgt, 0)

            output_dim = output.shape[-1]

            output = output[:, 1:].reshape(-1, output_dim)
            tgt = tgt[:, 1:].reshape(-1)

            loss = criterion(output, tgt)

            epoch_loss += loss.item()

    return epoch_loss / len(iterator)
```

```python
import tqdm
import numpy as np
import torch
import os
from torch.utils.data import DataLoader
checkpoint_dir = "/content/drive/MyDrive/AI_Project/checkpoints"
models_dir = "/content/drive/MyDrive/AI_Project/models"

n_epochs = 10
clip = 1.0
teacher_forcing_ratio = 0.5
best_valid_loss = float("inf")

best_model_path = f"{models_dir}/seq2seq_model.pt"

start_epoch = 0

resume_path = os.path.join(checkpoint_dir, "latest_checkpoint.pt")
if os.path.exists(resume_path):
    print(f"Loading checkpoint from {resume_path}")
    checkpoint = torch.load(resume_path,map_location=torch.device("cuda" if torch.cuda.is_available() else "cpu"))
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_epoch = checkpoint['epoch'] + 1
    best_valid_loss = checkpoint['best_valid_loss']
    print(f"Resuming from epoch {start_epoch}")

for epoch in tqdm.tqdm(range(start_epoch, n_epochs)):
    train_loss = train(
        model,
        train_iterator,
        optimizer,
        criterion,
        clip,
        teacher_forcing_ratio,
    )
    valid_loss = evaluate(
        model,
        valid_iterator,
        criterion,
    )

    if valid_loss < best_valid_loss:
        best_valid_loss = valid_loss
        torch.save(model.state_dict(), best_model_path)
        print(f"New best model saved to {best_model_path}")

    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_loss': train_loss,
        'valid_loss': valid_loss,
        'best_valid_loss': best_valid_loss
    }
    checkpoint_path = os.path.join(checkpoint_dir, "latest_checkpoint.pt")
    torch.save(checkpoint, checkpoint_path)

    if epoch % 2 == 0:
        numbered_checkpoint_path = os.path.join(checkpoint_dir, f"checkpoint_epoch_{epoch}.pt")
        torch.save(checkpoint, numbered_checkpoint_path)

    print(f"Epoch {epoch+1}/{n_epochs}")
    print(f"\tTrain Loss: {train_loss:.3f} | Train PPL: {np.exp(train_loss):.3f}")
    print(f"\tValid Loss: {valid_loss:.3f} | Valid PPL: {np.exp(valid_loss):.3f}")
```
